## LIST

‚úÖA list in Python is a collection of items (elements) that are ordered, changeable
(mutable), and allow duplicate elements.

Example:

    fruits = ["apple", "orange", "cherry", "apple"]
    
    print(fruits)
    
    Output: ['apple', 'orange', 'cherry', 'apple']

‚û°Ô∏èCreate List in Python

    You can create lists in Python by placing comma-separated values between square
    brackets []. Lists can contain elements of different data types, including other lists.

    Syntax: list_name = [element1, element2, element3, ...]

##### üîπList of strings
colors = ["red", "green", "blue"]

##### üîπList of integers
numbers = [1, 2, 3, 4, 5]

##### üîπMixed data types
mixed = [1, "hello", 3.14, True]

##### üîπNested list
nested = [1, [2, 3], [4, 5, 6]]

##### ‚û°Ô∏èList Traversing and methods

```python
number = [10,20,30,40,50]

# Add 100 to the list
number.append(100)
print(f"Append Method : {number}")
```

```python
number = [10,20,30,40,50]
#  insert Method
number.insert(2,1000)  #insert(index,value)
print(f"Insert Method : {number}")
```

```python
number = [10,20,30,40,50]
# add multiple elements at the end
number.extend([2,4,6,8]) #extend([a1,a2,a3,a4,....])
print(f"Extend Method : {number}")
```

```python
number = [10,20,30,40,50]
# removes the first occurance of 5
number.remove(10) 
print(f"Remove Method : {number}")
```

```python
number = [10,20,30,40,50]
# remove and store the element at index 3
number.pop(3) 
print(f"Pop Method : {number}")
```

üü®Print positive and negative elements of an List?

```python
lst = [1,5,3,-1,400,0,-33]
positive = []
negative = []
for i in lst:
    if(i<0):
        negative.append(i)
    else:
        positive.append(i)
print(f"Positive Elements : {positive}")
print(f"Negative Elements : {negative}")
```

üü®Mean of List elements?

```python
number = [1,2,3,4,5]
mean = 0
for i in number:
    mean+=i
print(f"Mean = {mean}")
```

üü®Find the greatest element and print its index too?

```python
lst = [1,200,3,4,5]
val = lst[0]
index = 0
for i in range(1,len(lst)):
   if(lst[i]>val):
       val=lst[i]
       index=i
print(f'Max Value : {val}')
print(f'Index Value : {index}')
```

üü®Find the second greatest element?

```python
lst = [1,200,3,4,5]
largest = lst[0]
sec_largest = float('-inf')
for i in lst:
    if i>largest:
        sec_largest=largest
        largest=i
    elif i>sec_largest and i!=largest:
        sec_largest=i
print(sec_largest, largest)

# Using Sort method
lst.sort()
lenght = len(lst)-2
print(lst[lenght])

```

üü®Check if List is sorted or not.

```python
lst = [1,2,3,4,5]
is_sorted = True
for i in range(len(lst)-1):
    if lst[i]>lst[i+1]:
        is_sorted = False
        break
if is_sorted:
    print("Sorted")
else:
    print("Not Sorted")
```
