ğŸŸ¨Print the sum of all even & odd numbers in a range
separately

```python
n = int(input("Enter Number : "))
even = 0
odd=0
for i in range(1,n+1):
    if(i%2==0):
        even+=i
    elif(i%2!=0):
        odd+=i
print(f"Even : {even}")
print(f"Odd : {odd}")
```

ğŸŸ¨Accept a number and check if it a perfect number or not.

```python
n = int(input("Enter Number : "))
sum_div = 0
for i in range(1,n):
    if(n%i==0):
        sum_div+=i
if(n==sum_div):
    print(f"{n} is a Perfect Number")
else:
    print(f"{n} is not a Perfect Number")
```

ğŸŸ¨Check wether the number is prime or not

```python
n = int(input("Enter Number : "))

if(n<=1):
    print(f"{n} is not a Prime Number.")
else:
    is_prime = True
    for i in range(2,int(n**0.5)+1):
        if n%i==0:
            is_prime = False
            break
    
    if is_prime:
        print(f"{n} is a Prime Number")
    else:
        print(f"{n} is not a Prime Number")
```

ğŸŸ¨Reverse a string without using in build functions.

```python
str_g = input("Enter String : ")
# we'll use not built-in fuctions , not even lenght {len}
rev = ''
# 1. find length of string first
length = 0
for _ in str_g:
    length+=1
    # print(length)

# 2. after finding the length , reverse the string.
i = length-1
while i>=0:
    rev+=str_g[i]
    i-=1
print(rev)
```

ğŸŸ¨Check string is Pallindrome or not

```python
n = int(input("Enter Nuber : "))
r_n=str(n)
rev_n = int(r_n[::-1])
print(rev_n)
if n==rev_n:
    print("Pallindrome")
else:
    print("not Pallindrome")
```
